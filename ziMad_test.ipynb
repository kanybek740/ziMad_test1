Считаем количество запуска игры

```python
launch = df[df['event_name'] == 'launch']
launch_count_perday = launch.groupby(['event_days'])['event_name'].agg(['count', 'nunique'])
launch_count_perday
```


Считаем сумму по дням

```python
value_count = df.groupby(['event_days'])['event_value'].agg(['sum','nunique'])
value_count
```

```python
arpdau = value_count['sum']/launch_count_perday['count']
arpdau
```
